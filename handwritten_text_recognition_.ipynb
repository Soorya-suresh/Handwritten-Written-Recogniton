## **HANDWRITTEN TEXT RECOGNITION USING TENSORFLOW**

Handwritten Text Recognition (HTR) system implemented using TensorFlow 2.x and trained on the IAM offline HTR datasets. This Neural Network model recognizes the text contained in the images of segmented texts lines.

Data partitioning (train, validation, test) was performed following the methodology of each dataset. The project implemented the HTRModel abstraction model (inspired by CTCModel) as a way to facilitate the development of HTR systems.
